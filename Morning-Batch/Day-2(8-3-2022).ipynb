```python
## Arihtmetic Operators
a = int(input("Enter a number: "))
b = int(input("Enter another number: "))
print(a+b,a-b,b*a,a/b,a//b,a**b,a%b,b%a)
```

## Assignment Operators

```python
k = 8

k += 5
print(k) # 13

k-=10
k*10
print(k) # 3

k %= 2
print(k) # 1

k += 20
k*=5
print(k) # 105
```

### Comparison Operators

```python
x = 20
y = 300
print(y>x,x<=y,x<y,x!=y,y==x,x>=y)
```

```python
x = 20
y = 10+10
print(y>x,x<=y,x<y,x!=y,y==x,x>=y)
```

```
Syntax:

    statements

    statements
```

````python
# positive or negative
num = int(input("Enter any value: "))
if num<0:
    print("-ve")
elif num>0:
    print("+ve")
else: 
    print(0)
````

### Nested if

```
if condition:
    statements
    if condition:
        statements
    else:
        statements
elif condition:
    statemtns
    if condition:
        statements
else:
    statements
    if condition:
        stetements
```

````python
# example
urn = 'vits@gmail.com'
pwd = 'Vit$890'
username = input("Enter username: ")
if urn==username:
    password = input("Enter ur password: ")
    if password==pwd:
        print("******** Welcome *********")
    else:
        print("Invalid Password!!!")
else:
    print("Invalid Username..")
````
